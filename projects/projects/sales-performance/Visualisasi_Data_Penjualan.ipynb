```python
import pandas as pd

# Membaca file Excel yang di-upload
df = pd.read_excel('dataset-penjualan.xlsx')

# Menampilkan beberapa baris pertama untuk memeriksa data
df.head()
```

1. PERSIAPAN DATA


```python
# Cek data yang hilang di kolom-kolom penting
missing_data = df[['Tanggal', 'Kode Barang', 'Nama Barang', 'Total Penjualan', 'Total Harga Pokok']].isnull().sum()
print(missing_data)
```

```python
df['Total Penjualan'] = df['Total Penjualan'].fillna(df['Jumlah'] * df['Harga Satuan'])
df['Total Harga Pokok'] = df['Total Harga Pokok'].fillna(df['Jumlah'] * df['Harga Pokok'])
```

```python
# Ubah kolom Tanggal menjadi tipe datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'], errors='coerce')

# Pastikan kolom numerik dalam format yang benar
df['Jumlah'] = pd.to_numeric(df['Jumlah'], errors='coerce')
df['Harga Satuan'] = pd.to_numeric(df['Harga Satuan'], errors='coerce')
df['Harga Pokok'] = pd.to_numeric(df['Harga Pokok'], errors='coerce')
```

```python
df = df.drop_duplicates()
```

```python
# Pastikan kolom angka dalam format numerik
df['Jumlah'] = pd.to_numeric(df['Jumlah'], errors='coerce')
df['Harga Satuan'] = pd.to_numeric(df['Harga Satuan'], errors='coerce')
df['Harga Pokok'] = pd.to_numeric(df['Harga Pokok'], errors='coerce')
df['Total Penjualan'] = pd.to_numeric(df['Total Penjualan'], errors='coerce')
df['Total Harga Pokok'] = pd.to_numeric(df['Total Harga Pokok'], errors='coerce')

# Ubah kolom Tanggal ke format datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'], errors='coerce')

# Buat kolom tambahan: Total Keuntungan
df['Total Keuntungan'] = df['Total Penjualan'] - df['Total Harga Pokok']
```

2. JAWABAN DARI PERTANYAAN BISNIS

a. Jumlah total transaksi (Record Count)

```python
jumlah_transaksi = df.shape[0]
print("Jumlah Transaksi:", jumlah_transaksi)
```

b. Jumlah barang yang terjual (Jumlah total dari Jumlah)

```python
# Total barang terjual (SUM semua kuantitas)
print("Total Barang Terjual (SUM):", df['Jumlah'].sum())

# Jumlah transaksi (jumlah baris / record count)
print("Jumlah Transaksi (COUNT):", len(df))

# Jumlah barang unik yang pernah terjual
print("Jumlah Barang Unik:", df['Nama Barang'].nunique())

# Cek ringkasan nilai Jumlah
print(df['Jumlah'].describe())
```

```python
# Ganti nama kolom
df_renamed = df.rename(columns={
    'Record Count': 'Jumlah Transaksi',
    'Nama Barang': 'Jumlah Barang'
})

# Cek hasilnya
df_renamed.head()
```

```python
# Buat kolom baru 'Jumlah Transaksi' yang isinya 1 per baris
df['Jumlah Transaksi'] = 1

# Rename kolom 'Nama Barang' jadi 'Jumlah Barang'
df = df.rename(columns={'Nama Barang': 'Jumlah Barang'})

# Cek hasil
df.head()
```

c. Jumlah total nilai penjualan (Total Penjualan)

```python
# Rename kolom dan tambahkan kolom transaksi
df['Jumlah Transaksi'] = 1
df = df.rename(columns={'Nama Barang': 'Jumlah Barang'})

# Simpan ke file CSV
df.to_csv('dataset_penjualan_updated.csv', index=False)
```

```python
total_penjualan = df['Total Penjualan'].sum()
print("Total Penjualan (Rp):", total_penjualan)
```

d. Jumlah total keuntungan (Total Keuntungan)




```python
total_keuntungan = df['Total Keuntungan'].sum()
print("Total Keuntungan (Rp):", total_keuntungan)
```

e. Komposisi transaksi berdasarkan kategori

```python
kategori_komposisi = df['Kategori'].value_counts()
print(kategori_komposisi)
```

f. Trend keuntungan per bulan

```python
df['Bulan'] = df['Tanggal'].dt.to_period('M')
keuntungan_bulanan = df.groupby('Bulan')['Total Keuntungan'].sum().reset_index()
print(keuntungan_bulanan)
```

g. kota dengan keuntungan terbesar

```python
keuntungan_per_kota = df.groupby('City')['Total Keuntungan'].sum().sort_values(ascending=False)
print(keuntungan_per_kota.head())
```

h. Kota dengan nilai penjualan terbesar

```python
penjualan_per_kota = df.groupby('City')['Total Penjualan'].sum().sort_values(ascending=False)
print(penjualan_per_kota.head())
```
